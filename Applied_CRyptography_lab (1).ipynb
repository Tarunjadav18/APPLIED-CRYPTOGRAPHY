**Implement prime number generation, GCD computation, and Euler’s Phi function?**

CODE 1A: Prime Number Generation

```python
# Program to generate prime numbers up to a given limit

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


limit = int(input("Enter the limit: "))
print("Prime numbers are:")

for num in range(2, limit + 1):
    if is_prime(num):
        print(num, end=" ")
```

CODE 1B: GCD Computation (Euclidean Algorithm)

```python
# Program to compute GCD of two numbers

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("GCD is:", gcd(a, b))
```

CODE 1C: Euler’s Phi (Totient) Function

```python
# Program to compute Euler's Phi Function

def euler_phi(n):
    result = n
    p = 2
    temp = n

    while p * p <= temp:
        if temp % p == 0:
            while temp % p == 0:
                temp //= p
            result -= result // p
        p += 1

    if temp > 1:
        result -= result // temp

    return result


n = int(input("Enter the value of n: "))
print("Euler Phi value:", euler_phi(n))
```

**Implement modular arithmetic operations and algebraic structures (groups, rings, and fields) for resource-constrained IoT devices?**

CODE 2A: Modular Arithmetic Operations

```python
# Program to perform Modular Arithmetic Operations

def mod_add(a, b, m):
    return (a + b) % m


def mod_sub(a, b, m):
    return (a - b) % m


def mod_mul(a, b, m):
    return (a * b) % m


def mod_exp(base, exp, m):
    result = 1
    base = base % m

    while exp > 0:
        if exp % 2 == 1:
            result = (result * base) % m
        base = (base * base) % m
        exp //= 2

    return result


# ---- MAIN ----
a = int(input("Enter a: "))
b = int(input("Enter b: "))
m = int(input("Enter modulus m: "))

print("Modular Addition:", mod_add(a, b, m))
print("Modular Subtraction:", mod_sub(a, b, m))
print("Modular Multiplication:", mod_mul(a, b, m))
print("Modular Exponentiation:", mod_exp(a, b, m))
```

**Simulate the Diffie–Hellman Key Exchange protocol for secure session key establishment**

```python
# Diffie–Hellman Key Exchange Simulation

def modular_exponentiation(base, exponent, modulus):
    """Efficient modular exponentiation"""
    result = 1
    base = base % modulus

    while exponent > 0:
        if exponent % 2 == 1:
            result = (result * base) % modulus
        base = (base * base) % modulus
        exponent //= 2

    return result


# -------- MAIN PROGRAM --------
# Publicly shared values
p = int(input("Enter a large prime number (p): "))
g = int(input("Enter a primitive root (g): "))

# Private keys (kept secret)
private_A = int(input("Enter private key of User A: "))
private_B = int(input("Enter private key of User B: "))

# Public keys generation
public_A = modular_exponentiation(g, private_A, p)
public_B = modular_exponentiation(g, private_B, p)

# Shared secret key computation
shared_key_A = modular_exponentiation(public_B, private_A, p)
shared_key_B = modular_exponentiation(public_A, private_B, p)

print("\n--- Diffie–Hellman Key Exchange ---")
print("Public Key of User A:", public_A)
print("Public Key of User B:", public_B)
print("Shared Secret Key (User A):", shared_key_A)
print("Shared Secret Key (User B):", shared_key_B)

if shared_key_A == shared_key_B:
    print("Secure session key established successfully!")
else:
    print("Key exchange failed!")
```
