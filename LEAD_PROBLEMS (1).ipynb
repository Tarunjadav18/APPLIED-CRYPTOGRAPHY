**Problem 1: Number Theory Foundations for Public-Key Cryptography**



Problem Statement 1
Design and implement a program that accepts two positive integers from the user and performs
the following operations to support public-key cryptographic computations:
1. Determine whether each number is a prime number.
2. Compute the GCD of the two numbers using the Euclidean Algorithm.
3. Check whether the two numbers are co-prime.
4. If the numbers are co-prime, compute Euler’s Phi Function for each number.
5. Using appropriate inputs, verify:
o Fermat’s Little Theorem
o Euler’s Theorem
6. Perform basic modular arithmetic operations (addition, multiplication, exponentiation)
relevant to cryptographic algorithms.


Objective:
To understand how prime numbers, GCD, co-primality, and number-theoretic theorems form the
mathematical foundation of public-key cryptography

```python
import math

# Check prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# GCD using Euclidean Algorithm
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Euler's Phi Function
def phi(n):
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result

# Modular Exponentiation
def mod_exp(a, b, m):
    result = 1
    a %= m
    while b > 0:
        if b % 2 == 1:
            result = (result * a) % m
        a = (a * a) % m
        b //= 2
    return result

# Main Program
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("\nPrime Check:")
print(a, "is Prime" if is_prime(a) else "is Not Prime")
print(b, "is Prime" if is_prime(b) else "is Not Prime")

g = gcd(a, b)
print("\nGCD =", g)

if g == 1:
    print("Numbers are Co-Prime")

    phi_a = phi(a)
    phi_b = phi(b)
    print("Phi(", a, ") =", phi_a)
    print("Phi(", b, ") =", phi_b)

    # Fermat's Little Theorem
    if is_prime(a):
        print("\nFermat's Little Theorem:")
        print(f"{b}^(a-1) mod {a} =", mod_exp(b, a-1, a))

    # Euler's Theorem
    print("\nEuler's Theorem:")
    print(f"{b}^phi({a}) mod {a} =", mod_exp(b, phi_a, a))

else:
    print("Numbers are NOT Co-Prime")

print("\nModular Arithmetic:")
print("(a + b) mod a =", (a + b) % a)
print("(a * b) mod a =", (a * b) % a)
print("b^2 mod a =", mod_exp(b, 2, a))
```


**Problem 2: Algebraic Structures and Modular Arithmetic in Cryptography**


Problem Statement 2
Develop a program that allows the user to define a finite set of integers under modulo arithmetic
and analyze its suitability for cryptographic applications by performing the following:
1. Verify whether the given set forms a Group, Ring, or Field under modular addition and
multiplication.
2. Validate algebraic properties such as closure, associativity, identity, inverse, and
distributivity.
3. Demonstrate how modular arithmetic operates within these algebraic structures.
4. Explain the importance of groups, rings, and fields in the design of public-key
algorithms such as RSA and ElGamal.


```python
import math

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def is_field(n):
    for i in range(1, n):
        if gcd(i, n) != 1:
            return False
    return True

n = int(input("Enter modulo value n: "))

print("\nSet Z_", n, "=", list(range(n)))

print("\nGroup under Addition:")
print("Closure ✔")
print("Associativity ✔")
print("Identity element: 0")
print("Inverse exists for all elements")

print("\nRing Properties:")
print("Z_", n, "is a Ring under addition and multiplication")

if is_field(n):
    print("\nZ_", n, "is a Field (n is Prime)")
else:
    print("\nZ_", n, "is NOT a Field (n is Composite)")

# Modular arithmetic demo
a, b = 3, 4
print("\nModular Arithmetic Demo:")
print("(3 + 4) mod", n, "=", (a + b) % n)
print("(3 * 4) mod", n, "=", (a * b) % n)

print("\nCryptographic Importance:")
print("Groups → Diffie-Hellman")
print("Rings → RSA")
print("Fields → Secure public-key encryption")
```
